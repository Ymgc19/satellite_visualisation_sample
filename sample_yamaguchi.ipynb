```python
import rasterio
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

# データの読み込み

```python
with rasterio.open('/content/drive/MyDrive/S1_VH_2024-02-16.tif') as src:
    vh20240216 = src.read()
with rasterio.open('/content/drive/MyDrive/S1_VH_2024-04-28.tif') as src:
    vh20240428 = src.read()
with rasterio.open('/content/drive/MyDrive/S1_VH_2024-12-12.tif') as src:
    vh20241212 = src.read()
```

```python
colors = ["blue", "blue", "yellow", "orange"]
custom_cmap = LinearSegmentedColormap.from_list("CustomCmap", colors)
```
